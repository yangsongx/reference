```python
objs = db.status.aggregate(
 [ 
   {'$match': {'host':'cnnx-prod-1'}},
   {'$project':{'used_cpu':1, 'used_mem':1, 'total_mem':1,'when':1, 'mywhen': {'$concat':[{'$dateToString': { 'format': "%m-%d %H:", 'date': "$when" }}, '$quarter_num']}}},
   {'$group': {'_id': '$mywhen', 'cpu_value': {'$avg':'$used_cpu'}}},
   {'$sort': {'when': 1}}
 ])
print(objs)
for it in objs:
    print(it)
```
